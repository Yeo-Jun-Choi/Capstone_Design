```python
X_speed['ans']
Y_speed['ans']
Z_speed['ans']
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(X_speed['ans']) # x축 각속도
data1 = pd.DataFrame(Y_speed['ans']) # y축 각속도
data2 = pd.DataFrame(Z_speed['ans']) # z축 각속도
```

```python
data = data.transpose()
data1 = data1.transpose()
data2 = data2.transpose()
```

```python
data[2]= data1[1] #y축 각속도
```

```python
data[3]= data2[1] #z축 각속도
```

```python
data['x_unit_vector'] = data[1]/(((data[1]**2)+(data[2]**2)+(data[3]**2))**0.5) #x축 각속도 단위벡터화
```

```python
data['y_unit_vextor'] = data[2]/(((data[1]**2)+(data[2]**2)+(data[3]**2))**0.5) #y축 각속도 단위벡터화
```

```python

data['z_unit_vector'] = data[3]/(((data[1]**2)+(data[2]**2)+(data[3]**2))**0.5) #z축 각속도 단위벡터화
```

```python
data = data[522:] #드론이 뜰때부터 자르기
```

```python
data.reset_index() #인덱스 초기화
```

```python
data = data.drop([data.columns[0]],axis=1)
```

```python
data["x_integer"] = data[1]*0.01
data["y_integer"] = data[2]*0.01
data["z_integer"] = data[3]*0.01
```

```python
data["x_integer"] = data["x_integer"].cumsum()
data["y_integer"] = data["y_integer"].cumsum()
data["z_integer"] = data["z_integer"].cumsum()
```
